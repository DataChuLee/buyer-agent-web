```python
from crawling import crazy11_info,redsoccer_info,soccerboom_info,cafo_info, crawl_info
from concurrent.futures import ThreadPoolExecutor, as_completed
from typing import List, Dict, Union, Optional
```

```python
# 코드 수정 후 -> 단일 사이트 30초 -> 11초 / 여러 사이트 -> 병렬로 설정하였으므로 동일하게 11초대 걸림
def crawl_info_parallel(
    site_keywords: List[str],
    product_keyword: str,
    min_price: int,
    max_price: int
) -> Dict[str, List[Dict]]:
    
    if not site_keywords:
        return {}

    deduped_sites = []
    seen_sites = set()
    for site in site_keywords:
        normalized = site.strip().lower()
        if normalized in seen_sites:
            continue
        seen_sites.add(normalized)
        deduped_sites.append(site)

    if len(deduped_sites) == 1:
        site = deduped_sites[0]
        try:
            return {site: crawl_info(site, product_keyword, min_price, max_price)}
        except Exception as e:
            return {site: [{"error": f"error: {str(e)}"}]}

    max_workers = min(4, len(deduped_sites))
    results = {}

    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        future_to_site = {
            executor.submit(crawl_info, site, product_keyword, min_price, max_price): site
            for site in deduped_sites
        }

        for future in as_completed(future_to_site):
            site = future_to_site[future]
            try:
                results[site] = future.result()
            except Exception as e:
                results[site] = [{"error": f"error: {str(e)}"}]
    return results
```

```python
result = crawl_info_parallel(["크레이지11","레드사커"], "나이키 머큐리얼 베이퍼", 100000, 199999)
```
